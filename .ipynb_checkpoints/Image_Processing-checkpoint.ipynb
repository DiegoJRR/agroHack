```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('field.png')   # you can read in images with opencv
img_hsv = cv.cvtColor(img, cv2.COLOR_BGR2HSV)

# hsv_color1 = np.asarray([166, 36, 34])   # white!
# hsv_color2 = np.asarray([174, 65, 19])   # yellow! note the order

# kernel = np.ones((5,5),np.float32)/25
# dst = cv.filter2D(img,-1,kernel)
dst = cv.blur(img, (5,5))
plt.imshow(dst)
plt.show()
```

```python
healthy_range = (np.asarray([ 42, 45, 13]), np.asarray([71,84,51]))   # white!   # yellow! note the order
warning_range = (np.asarray([99, 99, 85]), np.asarray([146, 166, 152]))   
# R: 85 G: 99 B: 99
# R: 152 G: 166 B: 146
healthy_mask = cv.inRange(dst, healthy_range[0], healthy_range[1])
warning_mask = cv.inRange(dst, warning_range[0], warning_range[1])

plt.imshow(healthy_mask, cmap='gray')   # this colormap will display in black / white
plt.show()
```

```python
result = cv2.bitwise_and(img, img, mask=healthy_mask)
# rgb(17, 122, 11)
img[np.where(healthy_mask)] = [11,122,17]
# rgb(226, 11, 11) 
img[np.where(warning_mask)] = [255,0,0]
plt.imshow(img)
plt.show()
```

```python
#Conteo de zona con cada rango de color
healthy_perc = np.sum(healthy_mask/255)/(img.shape[0]*img.shape[1])
warning_perc = np.sum(warning_mask/255)/(img.shape[0]*img.shape[1])
warning_perc
```
